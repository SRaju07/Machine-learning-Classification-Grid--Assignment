```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset['classification_yes'].value_counts()
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
independent.shape
```

```python
dependent=dataset[['classification_yes']]
```

```python
dependent.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
x_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
param_grid={"n_neighbors": [3, 5, 7, 11, 15],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]}
grid=GridSearchCV( KNeighborsClassifier(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring="accuracy")
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
```

```python
re
```

```python
format(grid.best_params_)
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
grid_predictions=grid.predict(x_test)
```

```python
grid_predictions
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_predictions)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)
```

```python
print(clf_report)
```
